```python
df = posting.dropna(subset=['Co_Nm'])
df = posting.dropna(subset=['Job_Desc'])
```

```python
df
```

```python
# Assuming 'Your_Column' is a categorical variable
distinct_values = df['Job_Ttl'].value_counts()

# Filter values with counts greater than 50
distinct_values_filtered = distinct_values[distinct_values > 50]
distinct_values_filtered = distinct_values_filtered.sort_values(ascending=False)

# Display distinct values
print("Distinct Values in 'Job_Ttl' with Counts Greater Than 50:")
print(distinct_values_filtered)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Job_Ttl', data=df[df['Job_Ttl'].isin(distinct_values_filtered.index)], order=distinct_values_filtered.index)
# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')
plt.title('Count of Distinct Values in Job_Ttl (Counts > 50)')
plt.show()
```
